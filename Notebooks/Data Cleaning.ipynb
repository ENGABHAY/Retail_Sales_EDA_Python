```python
import numpy as np
import pandas as pd
data = pd.read_csv("messy_retail_dataset.csv")
df=data.copy()
```

```python
df.head()
```

```python
df.info()
```

```python
#Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
#sort by orderID
df.sort_values(by="OrderID",ignore_index=True,inplace=True)
```

```python
#Converting into date formate
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
#Normalize the all text
df['ShippingMode']=df['ShippingMode'].str.title()
df["Product"]=df["Product"].apply(lambda x: x.title())
```

```python
 #filled Country null values using Region column
for i in df['Region'].dropna().unique():
    a = df[df["Region"] == i]["Country"]
    df.loc[df['Region'] == i, "Country"] = df.loc[df['Region'] == i, "Country"].fillna(a.mode()[0])
```

```python
 #filled Region null values using Country column
for i in df['Country'].dropna().unique():
    a = df[df["Country"] == i]["Region"]
    df.loc[df['Country']==i, "Region"] = df.loc[df['Country']==i, "Region"].fillna(a.mode()[0])
```

```python
#delete row where country and region are null
df.dropna(subset=["Region"], axis=0,how='any',inplace=True)
```

```python
#Replacing null values with zero assuming there not given
df["Discount"] = df["Discount"].fillna(0)
#Replacing with median for avoid bias
df["Profit"] = df["Profit"].fillna(df["Profit"].median())
```

```python
#Replacing null values into zero assuming No prfit and No Discount
df["Discount"].fillna(0, inplace=True)
df["Profit"].fillna(0, inplace=True)
```
